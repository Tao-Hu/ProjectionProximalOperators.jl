This example demonstrates the linear regression with max function penalty. Essentially, we are interested in solving the optimization problem of

$$ \mathrm{min} \frac{1}{2} \| y - X \beta \|_2^2 + \lambda \mathrm{max}(\beta) $$

The FISTA algorithm is demonstrated here to solve this problem. The key step of FISTA is to perform a proximal operator for max function in each iteration. In this implementation, I use the function *prox_max!* to perform the proximal operator for max function.

The data are simulated in the following way: let $X \in \mathtt{R}^{n \times p}$, where $n = 10000, p = 1000$. Each element of $X$ is simulated from standard normal. The true regression coefficients will be $\beta = (1,2,3,4,5,0,\ldots,0)^T$. Then generate response vector $y$ as: $y = X\beta$.

```julia
# create covariates matrix X and response vector y
srand(2015790003)
n = 10000
p = 1000
X = randn(n, p)
β = zeros(p)
k = 5
β[1:k] = 1:k
y = X * β;
```

## Solving the problem using FISTA algorithm over a grid

Load the source code and solve the problem with FISTA. For the grid of tunning parameter $\lambda$, I set it as $\lambda = 0,500,1000,\ldots,10000$.

```julia
include("src\\prox_max!.jl")
include("lrg_max.jl")
λList = [0:500:10000];
@time βhat_maxpath = lrgmaxpath(y, X, λList, alg = "FISTA");
```

Then plot the solution path along the grid.

```julia
using DataFrames, Gadfly

maxpath = DataFrame(Predictor = rep(1:p, length(λList)),
  Coeff = reshape(βhat_maxpath, (p * length(λList))),
  λ = rep(λList, each=p))

plot(maxpath, x =: λ, y =: Coeff, color=:Predictor, Geom.line,
  Guide.XLabel("λ"), Guide.YLabel("βhat(λ)"), 
  Guide.Title("Linear regression with max penalty"),
  Scale.color_discrete_hue)
```

According to the solution path, we can see that FISTA algorithm recover the true signal pretty well: the first five estimates are matched with true values and all rest coefficients are zeros. However, this solution path can not do variable selection compared with LASSO.

## Compared with *Convex.jl*

I have also tried to solve this proble using *Convex.jl* package. However, *Convex.jl* is still running after more than 20 minutes, while FISTA algorithm only takes about 1 seconds. Thus, we can sort of see the limit of current technology: for large scale problem, convex program solver may not be able to solve the problem in a proper time manner. By using "smart" algorithm, such as FISTA, and a good implementation, we should be able to conquer parts of the issues in big data era.
